```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import math
from PIL import Image
```

```python
img = Image.open('assets/lenna.png').convert('L')
```

# Preview a default image

```python
plt.imshow(img, cmap='gray')
```

```python
# convert image into a numpy array
img = np.asarray(img).astype('int32')
```

# Image encode

```python
def encode_image(image):
    result = np.zeros_like(image)
    for x in range(len(image)):
        for y in range(len(image[x])):
            if x-1 < 0 or y-1 < 0:
                result[x][y] = image[x][y]
                continue
            diff = (image[x-1][y] + image[x][y-1] - image[x-1][y-1])
            result[x][y] = image[x][y] - diff
            
    return result
```

```python
img_encoded = encode_image(img)
```

```python
plt.imshow(img_encoded, cmap='gray')
```

# Image decode

```python
def decode_image(image):
    result = np.zeros_like(image)
    for x in range(len(image)):
        for y in range(len(image[x])):
            if x-1 < 0 or y-1 < 0:
                result[x][y] = image[x][y]
                continue
            diff = (result[x-1][y] + result[x][y-1] - result[x-1][y-1])
            result[x][y] = image[x][y] + diff
            
    return result
```

```python
img_decoded = decode_image(img_encoded)
```

```python
plt.imshow(img_decoded, cmap='gray')
```
