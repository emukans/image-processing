```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
img = Image.open('assets/lenna.png').convert(mode='HSV')
```

```python
plt.imshow(img)
```

```python
img = np.array(img)
delta = 100
for i in range(img.shape[0]):
    for k in range(img.shape[1]):
        img[i][k][0] = img[i][k][0] + delta
        if img[i][k][0] > 360:
            img[i][k][0] -= 360
```

```python
img = Image.fromarray(img, 'HSV')
```

```python
plt.imshow(img)
```
